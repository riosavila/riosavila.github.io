```stata
*| echo: false
*| output: false
set scheme white2
color_style tableau
use data_slides/hotels-vienna.dta, clear
label var stars "Stars Rating N Stars"
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Absolute frequency

qui:histogram stars, d frequency ///
    scale(1.5) addlabels xlabel(1(.5)5)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Relative frequency
qui:histogram stars, d percent ///
    scale(1.5) addlabels xlabel(1(.5)5)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Price Distribution  1$ bin
qui:histogram price, d  ///
    scale(1.5) width(10) 
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Price Distribution 10$ bin
qui:histogram price,  ///
    scale(1.5) width(10)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Distance Distribution  
qui:histogram distance,   ///
    scale(1.5) 
```
