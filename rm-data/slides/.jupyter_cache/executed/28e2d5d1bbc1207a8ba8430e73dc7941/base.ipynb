```stata
*| echo: false
*| output: false
set scheme white2
color_style tableau
use data_slides/hotels-vienna.dta, clear
label var stars "Stars Rating N Stars"
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Absolute frequency

qui:histogram stars, d frequency ///
    scale(1.5) addlabels xlabel(1(.5)5)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Relative frequency
qui:histogram stars, d percent ///
    scale(1.5) addlabels xlabel(1(.5)5)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Price Distribution  1$ bin
qui:histogram price, d  ///
    scale(1.5) width(10) 
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Price Distribution 10$ bin
qui:histogram price,  ///
    scale(1.5) width(10)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Distance Distribution  
qui:histogram distance,   ///
    scale(1.5) 
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Default Bandwidth
 kdensity price,   ///
    scale(1.5)  note("") bw(10)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Bandwidth of 5
 kdensity price,   ///
    scale(1.5)  note("") bw(5)
```

```stata
*| echo: true
*| code-fold: true
*| fig-cap: Bandwidth of 20
 kdensity price,   ///
    scale(1.5)  note("") bw(20)
```

```stata
*| echo: false
*| fig-cap: "Box Plot: Viena prices"
*| fig-pos: center

use data_slides/hotels-vienna.dta, clear
drop if price>800

graph box price, title("Price distribution by stars") ///
  ytitle("Price in dollars (log Scale)") 
```

```stata
*| echo: false
*| fig-cap: "Box Plot: Viena prices"
*| fig-pos: center

use data_slides/hotels-vienna.dta, clear
drop if price>800
 graph box price if stars>1, ///
  over(stars)  xsize(10) ysize(6) ///
  ytitle("Price in dollars (log Scale)") ///
  title("Price distribution by stars") 
```
